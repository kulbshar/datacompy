Pivot Table

```python
import pandas as pd
```

```python
table2 = pd.read_csv('table2.csv')
```

```python
table2.info()
```

```python
table2.pivot_table(index=['Country'], columns= ['Medal'], values='Count',\
                  aggfunc=sum, fill_value=0)
```

```python
table2.groupby(['Country','Medal'])[['Count']].agg('sum').unstack().fillna(0)
```

```python
table2.pivot_table(index=['Country'], columns= ['Medal'], values='Count',\
                  aggfunc=sum, fill_value=0, margins=True, margins_name='Total')
```

cross tab

```python
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
pd.crosstab(titanic.sex, titanic.pclass)
```

```python
pd.crosstab(titanic.sex, titanic.pclass)
```

```python
titanic.groupby(['sex','pclass'])[['pclass']].agg('count').unstack()
```

```python
titanic.loc[:,['sex','pclass']].pivot_table(index='sex', \
                                            columns='pclass', values='pclass',\
                    aggfunc='count')
```

```python
pd.crosstab(titanic.sex, titanic.pclass,values=titanic.pclass, aggfunc='count')
```

```python
pd.crosstab(titanic.sex, titanic.pclass,\
            values=titanic.pclass,\
            aggfunc='count', margins=True)
```
