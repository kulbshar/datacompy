```python
import pandas as pd
import numpy as np
```

```python
def _read_excel_file(
    file_path,
    dtype=str,
    na_values=None,
    keep_default_na=False,
    skiprows=None,
    header=0,
    **kwargs,
):
    """Read in an Excel file, returns pd.DataFrame."""
    data_frame = pd.read_excel(
        io=file_path,
        dtype=dtype,
        na_values=na_values,
        keep_default_na=keep_default_na,
        skiprows=skiprows,
        header=header,
        **kwargs,
    )

    return data_frame
```

```python
def replace_field_name_spaces(df, regex='\s+', string="_", ):
    """Replace spaces in field names with supplied value

        Keyword Arguments:
            regex (str): regular expression pattern 
                        to be replace by string
            string {str} -- string to replace spaces
                            (default: {"_"})
    """
    df.columns = df.columns.str.replace(
        regex, string
    )
    return df

```

```python
def _read_csv_file(
    file_path,
    sep=",",
    dtype=str,
    na_values=None,
    keep_default_na=False,
    header=0,
    skiprows=None,
    **kwargs,
):
    """Read in a CSV file, returns pd.DataFrame."""
    data_frame = pd.read_csv(
        filepath_or_buffer=file_path,
        sep=sep,
        dtype=dtype,
        na_values=np.nan,
        keep_default_na=keep_default_na,
        header=header,
        skiprows=skiprows,
        **kwargs,
    )
    return data_frame
```
